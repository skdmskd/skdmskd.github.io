```python
# ============================================
# Trie 기반 영어 단어 검색 시스템
# ============================================

# 기능
# 1. Trie 자료구조
# 2. 단어 삽입
# 3. 단어 검색
# 4. 접두사 검색
# 5. 자동완성
# 6. 빈도수 기반 추천
# 7. words.txt 자동 로딩
# 8. BST와 속도 비교
# 9. 콘솔 UI
# ============================================

import time


# ============================================
# Trie Node
# ============================================

class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False
        self.frequency = 0


# ============================================
# Trie
# ============================================

class Trie:
    def __init__(self):
        self.root = TrieNode()
        self.word_count = 0

    # 단어 삽입
    def insert(self, word):
        node = self.root

        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()

            node = node.children[char]

        # 중복 단어 방지
        if not node.is_end:
            node.is_end = True
            self.word_count += 1

        node.frequency += 1

    # 단어 검색
    def search(self, word):
        node = self.root

        for char in word:
            if char not in node.children:
                return False

            node = node.children[char]

        return node.is_end

    # 접두사 검색
    def starts_with(self, prefix):
        node = self.root

        for char in prefix:
            if char not in node.children:
                return False

            node = node.children[char]

        return True

    # 자동완성
    def autocomplete(self, prefix, limit=10):
        node = self.root

        # prefix 위치까지 이동
        for char in prefix:
            if char not in node.children:
                return []

            node = node.children[char]

        words = []

        # DFS 탐색
        self._dfs(node, prefix, words)

        # 빈도수 기준 정렬
        words.sort(key=lambda x: x[1], reverse=True)

        return words[:limit]

    # DFS
    def _dfs(self, node, current_word, words):
        if node.is_end:
            words.append((current_word, node.frequency))

        for char, next_node in node.children.items():
            self._dfs(next_node, current_word + char, words)

    # 전체 단어 개수
    def total_words(self):
        return self.word_count


# ============================================
# BST 구현 (속도 비교용)
# ============================================

class BSTNode:
    def __init__(self, word):
        self.word = word
        self.left = None
        self.right = None


class BST:
    def __init__(self):
        self.root = None

    def insert(self, word):
        self.root = self._insert(self.root, word)

    def _insert(self, node, word):
        if node is None:
            return BSTNode(word)

        if word < node.word:
            node.left = self._insert(node.left, word)

        elif word > node.word:
            node.right = self._insert(node.right, word)

        return node

    def search(self, word):
        return self._search(self.root, word)

    def _search(self, node, word):
        if node is None:
            return False

        if node.word == word:
            return True

        if word < node.word:
            return self._search(node.left, word)

        return self._search(node.right, word)


# ============================================
# words.txt 로딩
# ============================================

def load_words(filename, trie, bst):
    count = 0

    try:
        with open(filename, "r", encoding="utf-8") as file:

            for line in file:
                word = line.strip().lower()

                # 영어 단어만
                if word.isalpha():

                    trie.insert(word)
                    bst.insert(word)

                    count += 1

        print(f"\n총 단어 수: {count}")
        print("Trie 생성 완료")

    except FileNotFoundError:
        print(f"{filename} 파일이 없습니다.")


# ============================================
# Trie vs BST 속도 비교
# ============================================

def speed_test(trie, bst, word):

    print("\n===== 속도 비교 =====")

    # Trie
    start = time.time()
    trie_result = trie.search(word)
    trie_time = time.time() - start

    # BST
    start = time.time()
    bst_result = bst.search(word)
    bst_time = time.time() - start

    print(f"\n검색 단어: {word}")

    print("\n[Trie]")
    print("결과:", trie_result)
    print(f"시간: {trie_time:.10f}초")

    print("\n[BST]")
    print("결과:", bst_result)
    print(f"시간: {bst_time:.10f}초")


# ============================================
# 메인 프로그램
# ============================================

def main():

    trie = Trie()
    bst = BST()

    # 단어 파일 로딩
    load_words("words.txt", trie, bst)

    print("\n===== 영어 단어 검색 시스템 =====")

    while True:

        print("\n1. 단어 검색")
        print("2. 접두사 검색")
        print("3. 자동완성")
        print("4. 속도 비교")
        print("5. 종료")

        choice = input("\n메뉴 선택: ")

        # ====================================
        # 단어 검색
        # ====================================

        if choice == "1":

            word = input("검색할 단어: ").lower()

            if trie.search(word):
                print(f"'{word}' 존재함")

            else:
                print(f"'{word}' 없음")

        # ====================================
        # 접두사 검색
        # ====================================

        elif choice == "2":

            prefix = input("접두사 입력: ").lower()

            if trie.starts_with(prefix):
                print(f"'{prefix}' 로 시작하는 단어 존재")

            else:
                print("없음")

        # ====================================
        # 자동완성
        # ====================================

        elif choice == "3":

            prefix = input("자동완성 prefix 입력: ").lower()

            results = trie.autocomplete(prefix)

            print("\n===== 자동완성 결과 =====")

            if results:

                for word, freq in results:
                    print(f"{word} (빈도수: {freq})")

            else:
                print("결과 없음")

        # ====================================
        # 속도 비교
        # ====================================

        elif choice == "4":

            word = input("검색 단어 입력: ").lower()

            speed_test(trie, bst, word)

        # ====================================
        # 종료
        # ====================================

        elif choice == "5":
            print("프로그램 종료")
            break

        else:
            print("잘못된 입력")


# ============================================
# 실행
# ============================================

if __name__ == "__main__":
    main()
```
